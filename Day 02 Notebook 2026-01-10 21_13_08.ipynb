#Filter smartphone based on price

```python
oct_events.select("product_id","brand","price").filter((oct_events.brand == 'samsung') & (oct_events.price > 800)).show(10)   
```

```python
Top_brands = oct_events.filter(oct_events.brand.isNotNull())\
.groupBy("brand").count().orderBy("count", ascending=False).limit(5)
Top_brands.show()
```

# Group Smartphone brands and sort by the count

```python
Top_brands = oct_events.filter(oct_events.brand.isNotNull()) \
    .groupBy("brand") \
    .count() \
    .orderBy("count", ascending=False) \
    .limit(5)
Top_brands.show()
```

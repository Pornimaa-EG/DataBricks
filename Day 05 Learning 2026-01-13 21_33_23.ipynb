#Import DeltaTable and Spark Functions

```python
from delta.tables import DeltaTable
from pyspark.sql.functions import *
from pyspark.sql.window import Window
```

```python
# Load Delta Table
delta_table = DeltaTable.forPath(spark,delta_path)
delta_table = DeltaTable.forName(spark,"events_table")
```

```python
%sql
DESCRIBE DETAIL EVENTS_TABLE
```

```python
%sql
DESCRIBE DETAIL DELTA_OCT_EVENTS
```

```python
# Deduplicate source to avoid multiple source row errors
window_spec = Window.partitionBy("user_session","event_time").orderBy(col("event_time").desc())

update_df = (oct_events
             .withColumn("rank", row_number().over(window_spec))
             .filter(col("rank") == 1)
             .drop("rank")
            )
```

##Count of rows - Before Merge

```python
# Before Merge - Count of Rows
before_Merge = spark.read.table("workspace.default.events_table").count()
print("Before Merge count of rows : ",before_Merge)
#Merge
delta_table.alias("t").merge(
    update_df.alias("s"),
    "t.user_session = s.user_session AND t.event_time = s.event_time")\
 .whenMatchedUpdateAll() \
 .whenNotMatchedInsertAll() \
 .execute()
```

##Count of rows - After Merge

```python
after_merge = spark.read.table("workspace.default.events_table").count()
print("After Merge count of rows : ",after_merge)
```
